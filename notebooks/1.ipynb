```python
import requests
import pandas as pd

# https://api.sunrise-sunset.org/json?lat=36.7201600&lng=-4.4203400

def get_sunrise_sunset(lat, lng, date):
    url = f"https://api.sunrise-sunset.org/json?lat={lat}&lng={lng}&date={date}"
    response = requests.get(url)
    
    if response.status_code == 200:
        data = response.json()
        return data['results']
    else:
        return None

TEST = get_sunrise_sunset(36.7201600, -4.4203400, "2024-01-01")  # Example coordinates for Malaga, Spain

df = pd.DataFrame.from_dict(TEST, orient='index', columns=['Value'])
print(df)
```

```python
import time

edit_df = edit_df.reset_index(drop=True)  # make sure indexes pair with number of rows

# empty df
Coord_df = pd.DataFrame(columns=['Country', 'Latitude', 'Longitude', 'Sunrise', 'Sunset','Day_length'])


Coord_list = []
i = -1
for index, row in edit_df.iterrows():
    i = i + 1
    time.sleep(1)
    # print(row['Country'], row['Latitude'], row['Longitude'])
    # Coord_list.append([row['Latitude'],row['Longitude']])
    sunrise_sunset = get_sunrise_sunset(row['Latitude'], row['Longitude'], 'today')
    if sunrise_sunset:
        Coord_list.append({
            'Country': row['Country'],
            'Latitude': row['Latitude'],
            'Longitude': row['Longitude'],
            'Sunrise': sunrise_sunset['sunrise'],
            'Sunset': sunrise_sunset['sunset']
        })
        print(Coord_list[i])
        
        
Coord_df = pd.DataFrame(Coord_list)
    
    
# print(Coord_list)
```

```python
edit_df = edit_df.reset_index(drop=True)  # make sure indexes pair with number of rows

# empty df
Coord_df = pd.DataFrame(columns=['Country','Date' 'Latitude', 'Longitude', 'Sunrise', 'Sunset','Day_length'])


Coord_list = []
i = -1
for index, row in edit_df.iterrows():
    
    time.sleep(1)
    # print(row['Country'], row['Latitude'], row['Longitude'])
    # Coord_list.append([row['Latitude'],row['Longitude']])
    
    for month in range(1, 13):
        i = i + 1
        print(month)
        sunrise_sunset = get_sunrise_sunset(row['Latitude'], row['Longitude'], f'2024-{month:02d}-01')
        if sunrise_sunset:
            Coord_list.append({
                'Country': row['Country'],
                'Date': f'2024-{month:02d}-01',
                'Latitude': row['Latitude'],
                'Longitude': row['Longitude'],
                'Sunrise': sunrise_sunset['sunrise'],
                'Sunset': sunrise_sunset['sunset'],
                'Day_length': sunrise_sunset['day_length']
            })
            print(Coord_list[i])
        
        
Coord_df = pd.DataFrame(Coord_list)
```
